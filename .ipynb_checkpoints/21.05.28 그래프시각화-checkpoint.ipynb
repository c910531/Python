```python
import matplotlib.pyplot as plt
data = [10, 5, 2, 6, 20, 15]
plt.plot(data)
plt.show() # 그래프만 나오게 하기
```

```python
import numpy as np
fig = plt.figure() 
ax1 = fig.add_subplot(2,2,1) # 좌상, 우상, 좌하 그래프 양식 위치하기
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.plot(np.random.randn(50).cumsum(), 'k--') # cumsum()은 'k--'는 검은점선을 그리기 위한 스타일 옵션
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(30).cumsum(), 'ko--', label = "rand 1")
ax.plot(np.random.randn(30).cumsum(), 'ro', label = "rand 2")
ax.legend(loc='best') # 범례
ax.set_xticks([0, 10, 20, 30]) # ticks 을 통해 x축 단위조정
ax.set_xticklabels(["start","stage1","stage2","stage3"], rotation=30) # x축 지정 가능
ax.set_xlabel("Progressive stage")
ax.set_yticks([-10, -5, 0, 5]) # y축 지정 가능
ax.set_ylabel("Value")
ax.set_title("Test Image")
#ax.text(10, -3, "first shock")
ax.annotate("first shock", xy=(13, -5), # 특정 지점 강조 표시 기능
           xytext=(14,-6),
           arrowprops=dict(facecolor="black"))
cir = plt.Circle((14,-6),2, color='b', alpha=0.3)
ax.add_patch(cir)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.2, 0.75), 0.5, 0.15, color='k', alpha=0.3) #(0.2,0.75 위치에), 0.4,0.15넓이의 사각형 출력
cir = plt.Circle((0.7,0.2),0.15, color='b', alpha=0.3) # 타원형으로 출력되는 이유는 캔버스가 정사각형이 아니기 때문
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color='g', alpha=0.5)
ax.text(0.3, 0.78, "This is a rectangle") # 글씨 출력
ax.text(0.6, 0.2, "This is a circle")
ax.text(0.15, 0.4, "This is a polygon")
ax.add_patch(rect)
ax.add_patch(cir)
ax.add_patch(pgon)
plt.show()
```
