```python
#    (3 )

dataA = np.random.standard_normal((100,2))
dataB = np.random.standard_normal((100,2))
dataC = np.random.standard_normal((100,2))
dataA += np.array((-3,-3))
dataB += np.array((3,4))
dataC += np.array((-1,4))
plt.scatter(dataA[:,0], dataA[:,1], color = "b")
plt.scatter(dataB[:,0], dataB[:,1], color = "0.5") # 0~1    
plt.scatter(dataC[:,0], dataC[:,1], color = "#A566FF")
plt.show()
```

```python
import matplotlib.cm as cm
N=256
angle = np.linspace(0, 8*2*np.pi, N)
radius = np.linspace(5, 1, N)
X = radius*np.cos(angle)
Y = radius*np.sin(angle)
plt.scatter(X, Y, c=angle, cmap=cm.Spectral) #cm.Greys, cm.hsv 
plt.show()
```

```python
X = np.linspace(-10, 10, 100)
Y1 = [2*x**2+1 for x in X]
Y2 = [3*x**2+5 for x in X]
Y3 = [0.5*x**2+3 for x in X]
plt.plot(X, Y1, linestyle="solid")
plt.plot(X, Y2, linestyle="dashed")
plt.plot(X, Y3, linestyle="dashdot")
plt.show()
```

```python
N=4
A = np.array([1, 2, 3, 3])
B = np.array([2, 3, 2, 1])
C = np.array([1, 4, 2, 3])
X = np.arange(N)
plt.bar(X, A, color="r")
plt.bar(X, A+B, color="b", bottom=A)
plt.bar(X, A+B+C, color = "g", bottom=A+B)
plt.show()
```

```python
N=4
A = np.array([1, 2, 3, 3])
B = np.array([2, 3, 2, 1])
C = np.array([1, 4, 2, 3])
X = np.arange(N)
plt.bar(X, A, color="r", hatch='/')
plt.bar(X, A+B, color="b", hatch='x', bottom=A) #hatch   
plt.bar(X, A+B+C, color = "g", hatch='\\', bottom=A+B)
plt.show()
```
